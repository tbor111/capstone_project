```python
categories = ['business', 'world', 'us', 'opinion']
```

```python
# get rid of any row with a minority category
data = data[data['section'].isin(categories)]
```

```python
text = "face toad freak"
word_tokenize(text)
```

```python
stem_list = []
stem_text = ''
ps = PorterStemmer()
for i in data['title']:
    word_list = word_tokenize(i)
    for i in word_list:
        ps.stem(i)
        stem_text = ''+i + ' '
        stem_list.append(stem_text)
```

```python
# fit a stemmer

stemmer = PorterStemmer()
```

```python
stem_titles = stemmer.fit_transform(data['title'])
```

```python
# vectorize titles
title_vec = CountVectorizer(ngram_range = (1,2), stop_words = 'english', strip_accents = 'ascii')
title_vec.fit(data['title'])
title_bigrams = pd.DataFrame(title_vec.transform(data['title']).todense(), columns = title_vec.get_feature_names())
```
