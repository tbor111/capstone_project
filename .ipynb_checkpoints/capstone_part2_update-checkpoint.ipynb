## Problem Statement

The goal of this project is to classify news articles with varying levels of specifity. The most basic task is to differentiate between opinion and non-opinion articles from a corpus of *New York Times* articles published in the winter of 2016/2017. If that is successful, I'll try to then classify articles into narrower topic classifications: world/US news, arts/fashion, sports, science/technology/health, and opinion. 

## Risks and Assumptions
Drawing all articles from a single news source at a specific point in time may reduce my model's ability to generalize. (I'm still considering other sources to augment the corpus.) 

## Data Acquisition

So far, I've amassed a corpus of 922 articles from the [*New York Times*](www.nytimes.com) using a webscraper built in Beautiful Soup. I'm storing title, author, section, date of publication, and article text in a local Postgres database and updating new titles daily.

```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm.session import sessionmaker
from bs4 import BeautifulSoup
import cookielib
import urllib2
import requests
import re
```

```python
def get_soup(url):
    #request = urllib2.Request(url)
    #response = urllib2.urlopen(url) #request
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html.decode('utf-8', 'ignore'), 'html.parser')
    results = soup.find_all('div', attrs = {'class': 'story-body'})
    print response.status_code
    return results

def scrape_soup(results):

    # initialize lists
    titles = []
    dates = []
    links = []
    full_texts = []
    authors = []
    sections = []

    # scrape results into lists
    for x in results:
        # get link
        link = x.find('a')['href']
        # set regex to eliminate interactive features
        match = re.search('^https://www.nytimes.com/20', link)
        if match:
            links.append(link)

    # resoup it
    for link in links:
        all_p = ''
        cj = cookielib.CookieJar()
        opener = urllib2.build_opener(urllib2.HTTPCookieProcessor(cj))
        new_soup = BeautifulSoup(opener.open(link).read().decode('utf-8', 'ignore'), 'html.parser')
        # get the article content
        body = new_soup.find_all('p', attrs = {'class': 'story-body-text story-content'})
        for p in body:
            new_p = p.text.strip()
            all_p = all_p + new_p
        full_texts.append(all_p)

        # get titles
        title = new_soup.find('meta', attrs = {'property': 'og:title'})['content']
        titles.append(title)

        # get authors
        author = new_soup.find('meta', attrs = {'name': 'author'})['content']
        authors.append(author)

        # get sections
        section = new_soup.find('meta', attrs = {'name': 'CG'})['content']
        sections.append(section)

        # get dates
        date = new_soup.find('meta', attrs = {'name': 'pdate'})['content']
        dates.append(date)

    data_dict = {
        'title': titles, 'link': links, 'author': authors, 'body': full_texts,
        'section': sections, 'date': dates
    }
    return data_dict

def make_df(data_dict):
    df = pd.DataFrame(data_dict)
    print "Adding {} rows to staging".format(len(df))
    return df
```

## Data Pipeline

Relying heavily on an [NLP tutorial](http://bbengfort.github.io/tutorials/2016/05/19/text-classification-nltk-sckit-learn.html), I've constructed a pipeline to preprocess article titles and text for use in different classifiers. I've experimented with NLTK's Naive Bayes, which requires data to be formatted in an entirely different way, but I'll probably end up preprocessing with NLTK and building models in Scikit-Learn. This pipeline takes in raw text, makes everything lowercase, strips punctuation, lemmatizes, removes stopwords, and vectorizes using TF-IDF. Data is then ready to use in the def_and_evalutate function. Currently, I'm using Multinomial Naive Bayes in Scikit-Learn, but I will likely experiment with additional estimators and compare performance. The pipeline also "pickles" the trained model, saving it for future use without the time of re-training. (This is useful due to the time demands of processing large bodies of text).

```python
from sklearn.base import BaseEstimator, TransformerMixin
from nltk import WordNetLemmatizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
import pickle
from nltk import sent_tokenize, pos_tag, wordpunct_tokenize
from nltk.corpus import wordnet as wn
```

```python
def identity(arg):
    return arg
```

```python
# make preprocessor class
class NLTKPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self, stopwords = None, punct = None, lower = True, strip = True):
        self.lower = lower
        self.strip = strip
        self.stopwords = set(nltk.corpus.stopwords.words('english'))
        self.punct = set(string.punctuation)
        self.lemmatizer = WordNetLemmatizer()
        
    def fit(self, X, y=None):
        return self
    def inverse_transform(self, X):
        return [" ".join(doc) for doc in X]
    def transform(self, X):
        return[
            list(self.tokenize(doc)) for doc in X
        ]
    def tokenize(self, document):
        document = document.encode('ascii', errors = 'replace')
        #document = ''.join([ch.lower() for ch in text if ch not in string.punctuation])
        for sent in sent_tokenize(document):
            # Break the sentence into part of speech tagged tokens
            for token, tag in pos_tag(wordpunct_tokenize(sent)): 
                token = token.lower() if self.lower else token
                token = token.strip() if self.strip else token
                token = token.strip('_') if self.strip else token
                token = token.strip('*') if self.strip else token
                if token in self.stopwords:
                    continue
                if all(char in self.punct for char in token):
                    continue
                lemma = self.lemmatize(token, tag)
                yield lemma
                
    def lemmatize(self, token, tag):
        tag = {
            'N': wn.NOUN,
            'V': wn.VERB,
            'R': wn.ADV,
            'J': wn.ADJ
        }.get(tag[0], wn.NOUN)
        return self.lemmatizer.lemmatize(token, tag)
```
